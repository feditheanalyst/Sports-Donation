```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
sport = pd.read_csv("./sport donations.csv")
sport.head()
```

```python
sport.isna().sum()
```

```python
sport.dropna(inplace= True)
```

```python
sport.head()
```

```python
league = sport.groupby(["League"])
a = league["Amount"].sum().rename_axis("league").reset_index(name = "amount")
a 
```

```python
a.sort_values(by= "amount",ascending= False, inplace=True)
```

```python
a
```

```python
b = a["amount"].tolist()
c = a["league"].tolist()
```

```python
plt.style.use("fivethirtyeight")
plt.barh(c, b, color = "black")
plt.rcParams["figure.figsize"] = (12,6)
plt.title("Total Donations For Each League")
plt.xlabel("Amount")
plt.tight_layout()
for i, v in enumerate(b):
    plt.text(v + 1, i + .000000, "$" + str(v), color = "black", fontweight = "bold")
plt.show();
```

```python
owner = sport.groupby(["Owner"])
d = owner["Amount"].sum().rename_axis("Owner").reset_index(name = "amount")
d 
```

```python
d.sort_values(by= "amount",ascending= False, inplace=True)
e = d.head(10)
e 
```

```python
f = e["Owner"].tolist()
g = e["amount"].tolist()
```
